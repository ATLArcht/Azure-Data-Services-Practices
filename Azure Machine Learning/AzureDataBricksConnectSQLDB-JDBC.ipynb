```
# create SQL DB and get SQL DB credential
jdbcHostname = "****.database.windows.net"
jdbcDatabase = "AdventureWorks2017"
jdbcPort = 1433
username = "****"
password ="*****"
jdbcUrl = "jdbc:sqlserver://{0}:{1};database={2};user={3};password={4}".format(jdbcHostname, jdbcPort, jdbcDatabase, username, password)

print(jdbcUrl)
```

```
# connection for JDBC
connectionProperties = {
  "user" : username,
  "password" : password,
  "driver" : "com.microsoft.sqlserver.jdbc.SQLServerDriver"
}
```

```
# run the push down SQL DB TSQL query, note use alias with () for the T-SQL query
pushdown_query = "(select * from [dbo].[MyViewEmployeeInfo]) myquery"
df = spark.read.jdbc(url=jdbcUrl, table=pushdown_query, properties=connectionProperties)
display(df)
```
