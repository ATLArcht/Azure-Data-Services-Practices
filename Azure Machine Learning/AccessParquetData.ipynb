```python
from azureml.core import Dataset, Datastore
from azureml.data.datapath import DataPath
from azureml.core import Workspace
```

```python
# get worksapce
ws = Workspace.from_config()
```

```python

   # create tabular dataset from Parquet files in datastore
   datastore = Datastore.get(ws, 'workspaceblobstore')
   datastore_path = [
       DataPath(datastore, 'UI/09-29-2020_054112_UTC/MyParquetIris.parquet')
   ]
   tabular = Dataset.Tabular.from_parquet_files(path=datastore_path)
```

```python
   # move data to pandas dataframe and display data 
   df = tabular.to_pandas_dataframe()
   print (df)
```
