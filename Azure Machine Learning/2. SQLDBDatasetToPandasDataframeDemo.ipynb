Convert tabular dataset in SQL DB datastore to Pandas dataframe

https://docs.microsoft.com/en-us/python/api/azureml-core/azureml.data.dataset_factory.tabulardatasetfactory?view=azure-ml-py#from-sql-query-query--validate-true--set-column-types-none--query-timeout-30-

```python
from azureml.core import Dataset, Datastore
from azureml.data.datapath import DataPath

# get workspace
from azureml.core import Workspace
ws = Workspace.from_config()

# create tabular dataset from a SQL database in datastore
datastore = Datastore.get(ws, 'sqldbdatastore')
query = DataPath(datastore, 'SELECT * FROM MyViewEmployee')
tabular = Dataset.Tabular.from_sql_query(query, query_timeout=10)
df = tabular.to_pandas_dataframe()

print(df)
```
