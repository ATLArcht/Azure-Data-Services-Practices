Get experiment list 

```python
import azureml.core
from azureml.core import Workspace
from azureml.core.experiment import Experiment

ws = Workspace.from_config()

experiments = Experiment.list(ws)
for experiment in experiments:
    print(experiment.name)
```

Get published piepline list and ID

```python
from azureml.pipeline.core import Pipeline, PublishedPipeline

published_pipelines = PublishedPipeline.list(ws)
for published_pipeline in  published_pipelines:
    print(f"{published_pipeline.name},'{published_pipeline.id}'")
```

Create a time based Schedule

refer ScheduleRecurrence SDK
https://docs.microsoft.com/en-us/python/api/azureml-pipeline-core/azureml.pipeline.core.schedulerecurrence?view=azure-ml-py

```python
from azureml.pipeline.core.schedule import ScheduleRecurrence, Schedule

MyRecurrence = ScheduleRecurrence(frequency="Month", interval=1)
recurring_schedule = Schedule.create(ws, name="MyRecurringSchedule906", 
                            description="Based on time",
                            pipeline_id=pipeline_id, 
                            experiment_name=experiment_name, 
                            recurrence=MyRecurrence)
```
